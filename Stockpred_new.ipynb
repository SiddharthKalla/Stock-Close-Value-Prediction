```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Processed_DJI.csv')
```

```python
final_data = data.fillna(data.median())
```

```python
final_data
```

```python
x = final_data.drop('Date', axis = 1)
x = x.drop('Name', axis = 1)
```

```python
x
```

```python
y = final_data.iloc[:,1].values
```

```python
y.shape
```

```python
x_train = x.iloc[:1500, :].values
x_test = x.iloc[1500:, :].values
y_train = y[:1500]
y_test = y[1500:]
```

```python
x_train
```

```python
X_train = []
Y_train = []
for i in range(60, 1500):
    X_train.append(x_train[i-60:i,:])
    Y_train.append(y_train[i])
X_train, Y_train = np.array(X_train), np.array(Y_train)
```

```python
X_train = np.expand_dims(X_train, axis = 3)
```

```python
X_train.shape
```

```python
import tensorflow as tf
```

```python
model = tf.keras.Sequential(
    [
        tf.keras.layers.InputLayer(input_shape = (60,82,1)),                                          
        tf.keras.layers.Conv2D(8, (1,82), activation = "linear"),
        tf.keras.layers.Conv2D(8, (3,1), activation = "linear"),            
        tf.keras.layers.MaxPooling2D(pool_size = (2,1)),                                  
        tf.keras.layers.Conv2D(8, (3,1), activation = "linear"),                                                             
        tf.keras.layers.MaxPooling2D(pool_size = (2,1)),                                                                
        tf.keras.layers.Flatten(),                                                         
        tf.keras.layers.Dense(1,),                                                                                           
    
    ]
)
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model_history = model.fit(X_train, Y_train, batch_size = 64, epochs = 100,verbose = 1)
```
