```python
training_data = data_final[data_final.index < 1487]
testing_data = data_final[data_final.index >= 1487]
```

```python
training_data.shape
```

```python
scaled_train_data = scaler.fit_transform(training_data)
scaled_test_data = scaler.fit_transform(testing_data)
```

```python
scaler_y = MinMaxScaler(feature_range = (0,1))
```

```python
scaled_train_data
```

```python
scaled_train_data.shape
```

```python
X_train = []
Y_train = []

for i in range(60, scaled_train_data.shape[0]-60):
    X_train.append(scaled_train_data[i-60:i])
    Y_train.append(scaled_train_data[i,0])
```

```python
X_test = []
Y_test = []

for i in range(60, scaled_test_data.shape[0]-60):
    X_test.append(scaled_test_data[i-60:i])
    Y_test.append(scaled_test_data[i,0])
```

```python
X_test, y_test = np.array(X_test), np.array(Y_test)
```

```python
y_test = np.expand_dims(Y_test, axis = 1)
```

```python
scaler_y.fit_transform(y_test)
```

```python
X_train, y_train = np.array(X_train), np.array(Y_train)
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.optimizers import Adam, RMSprop
from sklearn.metrics import mean_squared_error
from keras import metrics
```

```python
model = Sequential()

model.add(Bidirectional(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 72))))
model.add(Dropout(0.2))

model.add(Bidirectional(LSTM(units = 50, return_sequences = True)))
model.add(Dropout(0.2))

model.add(Bidirectional(LSTM(units = 50, return_sequences = True)))

model.add(Bidirectional(LSTM(units = 50)))
model.add(Dropout(0.2))

model.add(Dense(units = 1))
```

```python
from keras.utils.vis_utils import plot_model
plot_model(model, to_file = 'model.png', show_shapes=True)
```

```python
model.compile(optimizer = Adam(), loss = 'mean_squared_error')
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

```python
metrics = pd.DataFrame(model.history.history)
```

```python
metrics
```

```python
metrics[['loss','val_loss']].plot(figsize=(12, 5))
plt.title("Loss Vs Val_loss")
plt.xlabel('Epochs')
plt.ylabel('Loss')
```

```python
loss = model.evaluate(X_test, y_test)
loss
```

```python
pred = model.predict(X_test)
Final_pred = scaler_y.inverse_transform(pred)
```

```python
scale_close = 1/5.00575662e-03
```
